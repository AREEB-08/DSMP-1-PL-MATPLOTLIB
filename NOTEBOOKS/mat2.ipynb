```python
import matplotlib.pyplot as plt

# Set overall figure size correctly using `figsize` in subplots, not plt.figure()
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15, 7.5))

# Plot 1: avg vs strike_rate
ax[0][0].scatter(bf['avg'], bf['strike_rate'], color='blue', label='Avg vs SR')
ax[0][0].set_title('Avg vs Strike Rate', fontsize=12)
ax[0][0].set_xlabel('Average', fontsize=10)
ax[0][0].set_ylabel('Strike Rate', fontsize=10)
ax[0][0].legend()

# Plot 2: avg vs runs
ax[0][1].scatter(bf['avg'], bf['runs'], color='green', label='Avg vs Runs')
ax[0][1].set_title('Avg vs Runs', fontsize=12)
ax[0][1].set_xlabel('Average', fontsize=10)
ax[0][1].set_ylabel('Runs', fontsize=10)
ax[0][1].legend()

# Plot 3: strike_rate vs runs
ax[1][0].scatter(bf['strike_rate'], bf['runs'], color='red', label='SR vs Runs')
ax[1][0].set_title('Strike Rate vs Runs', fontsize=12)
ax[1][0].set_xlabel('Strike Rate', fontsize=10)
ax[1][0].set_ylabel('Runs', fontsize=10)
ax[1][0].legend()

# Plot 4: strike_rate vs avg
ax[1][1].scatter(bf['strike_rate'], bf['avg'], color='purple', label='SR vs Avg')
ax[1][1].set_title('Strike Rate vs Avg', fontsize=12)
ax[1][1].set_xlabel('Strike Rate', fontsize=10)
ax[1][1].set_ylabel('Average', fontsize=10)
ax[1][1].legend()

# Add spacing to prevent overlap
plt.tight_layout()

# Show all plots
plt.show()
```

```python
fg=plt.figure(figsize=(15,7))
ax=plt.subplot(projection='3d')
ax.scatter3D(bf['strike_rate'], bf['avg'],bf['runs'],color='green')
ax.set_xlabel('STRIKE RATE')
ax.set_ylabel('AVERAGE')
ax.set_zlabel('RUNS')
```

```python
dl=pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
dl['ballnumber'].unique()
sb=dl[dl['ballnumber'].isin([ 1,  2,  3,  4,  5,  6]) & (dl['batsman_run']==6)]
```

```python
grid=sb.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
grid
```

```python
plt.figure(figsize=(15,7))
plt.imshow(grid)

plt.colorbar()
```
