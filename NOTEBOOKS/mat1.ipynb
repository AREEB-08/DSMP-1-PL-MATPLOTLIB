```python
b=pd.read_csv('batsman_season_record.csv')
b
```

```python
x = np.arange(len(b['batsman']))  # label locations
width = 0.25  # width of each bar

# Create the plot
plt.bar(x - width, b['2015'], width, label='2015', color='skyblue')
plt.bar(x, b['2016'], width, label='2016', color='lightgreen')
plt.bar(x + width, b['2017'], width, label='2017', color='salmon')

# Labels & formatting
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.title('Runs by Batsmen in 2015, 2016, and 2017')
plt.xticks(x, b['batsman'], rotation=45)
plt.legend()
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```
